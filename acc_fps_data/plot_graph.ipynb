```python
import seaborn as sns
import matplotlib.pyplot as plt
import gzip
import pickle
import numpy as np
```

# acc metric graph plot
- Using first Lite-HRNet Result + Previous Frame joints
- Using only Previous Frame Joints

```python
with gzip.open('acc_first_prev.pkl', "rb") as f: 
    acc_first_prev = pickle.load(f)
acc_first_prev = acc_first_prev.tolist()
acc_first_prev
```

```python
with gzip.open('acc_only_prev.pkl', "rb") as f: 
    acc_only_prev = pickle.load(f)
acc_only_prev = acc_only_prev.tolist()
acc_only_prev
```

```python
x = np.arange(2, 11, 1)

sns.lineplot(x, acc_first_prev, label = "using lite-hrnet result and encoder result", color = "blue", alpha = 2)
sns.lineplot(x, acc_only_prev, label = "using only encoder result", color = "red", alpha = 2)

plt.xlabel("N (hyperparameter)")
plt.ylabel("average pixel distance")

plt.title("Average pixel distance btw lite-hrnet result and encoder result", fontsize = 15)

plt.grid(True, alpha = 0.3)
plt.show()

```

# FPS Metric graph plot
- Only Lite-HRNet
- Using first Lite-HRNet Result + Previous Frame joints
- Using only Previous Frame Joints

```python
with gzip.open('fps_first_prev.pkl', "rb") as f: 
    fps_first_prev = pickle.load(f)
fps_first_prev = fps_first_prev.tolist()
fps_first_prev
```

```python
with gzip.open('fps_hrnet.pkl', "rb") as f: 
    fps_hrnet = pickle.load(f)
fps_hrnet
```

```python
with gzip.open('fps_only_prev.pkl', "rb") as f: 
    fps_only_prev = pickle.load(f)
fps_only_prev = fps_only_prev.tolist()
fps_only_prev
```

```python
x = np.arange(2, 11, 1)

sns.lineplot(x, fps_first_prev, label = "using lite-hrnet result and encoder result", color = "#119AE8", alpha = 2)
sns.lineplot(x, fps_only_prev, label = "using only encoder result", color = "#A2D5F2", alpha = 2)
sns.lineplot(x, fps_hrnet, label = "Average FPS using only Lite-HRNet", color = "red", alpha = 2)


plt.xlabel("N (hyperparameter)")
plt.ylabel("FPS")

plt.title("FPS result", fontsize = 15)

plt.legend(bbox_to_anchor = (1, 0.5))
plt.grid(True, alpha = 0.3)
plt.show()
```
